```python
def get_link(articleUrl):
    html = urlopen("https://en.wikipedia.org" + articleUrl)
    bsObj = BeautifulSoup(html, 'lxml')
    return bsObj.find('div', {'id': 'bodyContent'}).findAll('a', href=re.compile('^(/wiki/)(.(?!:))*$'))
```

# 案例实战三

获取词条以及关联词条的url、标题、概要

```python
pages = set()

def get_links(pageUrl):
    global pages
    html = urlopen("https://en.wikipedia.org" + pageUrl)
    bsObj = BeautifulSoup(html, 'lxml')
    try:
        # 维基词条页面的特征：
        # 1、所有的标题都在h1->span标签里，而且只有一个标题标签
        # 2、正文文字在div#bodyContent标签里；第一段文字 div#mw-content-text->p
        # 3、编辑链接只出现在词条页面上，位于li#ca-edit标签里的 li#caedit->span->a
        print("https://en.wikipedia.org" + pageUrl)
        print(bsObj.h1.get_text())
        print(bsObj.find(id='mw-content-text')).findAll('p')[0]
        print(bsObj.find(id='ca-edit')).find('span').find('a').attr['href']
    except:
        print("This page is missing something! No worries though!")
    
    allLinks = bsObj.find_all('a', href=re.compile("^(/wiki/)((?!:).)*$"))
    for link in allLinks:
        if 'href' in link.attrs:
            newPage = link.attrs['href']
            print("------------------------\n" + newPage)
            pages.add(newPage)
            getLinks(newPage)

getLinks("/wiki/Farouk_Topan")
```

# 案例实战四

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re

random.seed(datetime.datetime.now())
allExtLinks = set()
allIntLinks = set()

# 获取页面所有内链的列表
def getInternalLinks(bsObj, includeUrl):
    internalLinks = []
    print('includeUrl: ', includeUrl)
    # Finds all links that begin with a "/"
    for link in bsObj.findAll('a', href=re.compile('^/.*')):
        if link.attrs['href'] is not None:
            if link.attrs['href'] not in internalLinks:
                internalLinks.append(link.attrs['href'])
    return internalLinks

# 获取所有外链的列表
def getExternalLinks(bsObj, excludeUrl):
    externalLinks = []
    print('externalUrl: ', excludeUrl)
    # Finds all links that start with "http" or "www" that do not contain the current URL
    for link in bsObj.findAll('a', href=re.compile('^(http|www)((?!' + excludeUrl + ').)*$')):
        if link.attrs["href"] is not None:
            if link.attrs['href'] not in externalLinks:
                externalLinks.append(link.attrs['href'])
    return externalLinks

# url拆分
def splitAddress(address):
    addressParts = address.replace('http://', '').split('/')
    return addressParts

# Collects a list of all external URLs found on the site
def getAllExternalLinks(siteUrl):
    html = urlopen(siteUrl)
    bsObj = BeautifulSoup(html, 'lxml')
#     print(splitAddress(siteUrl)[0])
    internalLinks = getInternalLinks(bsObj, splitAddress(siteUrl)[0])
    print(internalLinks)
    externalLinks = getExternalLinks(bsObj, splitAddress(siteUrl)[0])
    
    for link in externalLinks:
        if link not in allExtLinks:
            allExtLinks.add(link)
            print('link: ', link)
    for link in internalLinks:
        if link not in allIntLinks:
            allIntLinks.add(link)
            getAllExternalLinks("http:" + link)
            
            
getAllExternalLinks("http://oreilly.com")
```
