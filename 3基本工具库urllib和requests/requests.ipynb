## 实例一：京东商品信息的爬取

```python
import requests
```

```python
url = r'https://item.jd.com/100006382674.html'
try:
    r = requests.get(url, timeout=30)
    r.raise_for_status()
    r.encoding = r.apparent_encoding
    print(r.text[:2000])
except:
    print('fail')
```

## 实例二：亚马逊商品信息的爬取

```python
import requests
```

```python
url = 'http://www.amazon.cn/gp/product/B01M8L5Z3Y'
try:
    kv = {'use_agent': 'Mozilla/5.0'}
    r = requests.get(url, headers=kv)
    r.raise_for_status
    r.encoding = r.apparent_encoding
    print(r.text[1000:2000])
except:
    print('fail')
```

## 实例三：百度搜索关键字提交

```python
import requests
```

```python
keyword = 'python'
try:
    kv = {'wd': keyword}
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.84 Safari/537.36"}
    r = requests.get('https://www.baidu.com/s', params=kv, headers=headers)
    r.raise_for_status
    r.encoding = r.apparent_encoding
    print(r.text)
except:
    print('fail')
```

## 实例四：网络图片的爬取

```python
import requests
import os
```

```python
try:
    url = "https://odonohz90.qnssl.com/library/145456/bb0b3faa7a872d012bb4c57256b47585.jpg?imageView2/2/w/1000/h/1000/q/75"
    root = 'pic'
    path = root + url.split('/')[-1]
    if not path.endswith('.jpg'):
        path += '.jpg'
    if not os.path.exists(root):
        os.mkdir(root)
    if not os.path.exists(path):
        r = requests.get(url)
        f = open(path, 'wb')
        f.write(r.content)
        f.close()
        print(path)
    else:
        print(path, ' existed')
except:
    print('fail')
```
