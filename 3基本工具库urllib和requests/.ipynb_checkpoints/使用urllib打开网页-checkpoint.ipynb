## 最基本的方法打开网页

```python
from urllib.request import urlopen
```

```python
response = urlopen('http://www.baidu.com')
print(type(response))
print(response.status)
print(response.getheaders())
print(response.getheader('Server'))
html = response.read()
print(html[:2000])
```

## 携带data参数打开网页

```python
from urllib.parse import urlencode
from urllib.request import urlopen
```

```python
data = bytes(urlencode({'world': 'hello'}), encoding='utf8')
response = urlopen('http://httpbin.org/post', data=data)
print(response.read().decode('utf8'))
```

## 携带timeout参数打开网页1

```python
from urllib.request import urlopen
```

```python
response = urlopen('http://httpbin.org/get', timeout=1)
print(response.read().decode('utf8'))
```

## 携带timeout参数打开网页2

```python
from urllib.request import urlopen
```

```python
try:
    response = urlopen('http://httpbin.org/get', timeout=0.1)
    print(response.read())
except Exception as e:
    print(e)
```

## 通过构建Request打开网页1

```python
from urllib.request import Request
from urllib.request import urlopen
```

```python
request = Request('https://python.org')
response = urlopen(request)
print(response.read().decode('utf8'))
```

## 通过构建Request打开网页2

```python
from urllib.request import Request
from urllib.request import urlopen
from urllib.parse import urlencode
```

```python
url = 'http://httpbin.org/post'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36',
    'Host': 'httpbin.org'
}
dict = {'name': 'Germey'}
data = bytes(urlencode(dict), encoding='utf8')
req = Request(url=url, data=data, headers=headers, method='POST')
response = urlopen(req)
print(response.read().decode('utf8'))
```

## 与通过构建Request打开网页2对比

```python
from urllib.request import Request
from urllib.request import urlopen

req = Request(url=url, data=data, method='POST')
response = urlopen(req)
print(response.read().decode('utf-8'))
```

## 通过构建Request打开网页3：通过add_header方

```python
from urllib.request import Request
from urllib.request import urlopen
from urllib.parse import urlencode

url = 'http://httpbin.org/post'
dict = {'name': 'Germey'}
data = bytes(urlencode(dict), encoding='utf8')
req = Request(url=url, data=data, method='POST')
req.add_header('User-Agent', 'Mozilla/4.0(compatibe;MSIE 5.5;Windows NT)')
response = urlopen(req)
print(response.read().decode('utf-8'))
```

## 使用代理打开网页

```python
from urllib.error import URLError
from urllib.request import ProxyHandler, build_opener
```

```python
proxy_handler = ProxyHandler({'http': '106.56.102.140:8070'})
opener = build_opener(proxy_handler)
try:
    response = opener.open('http://www.baidu.com/')
    print(response.read().decode('utf8'))
except URLError as e:
    print(e.reason)
```
